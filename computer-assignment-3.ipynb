## 3.17

#### S1 is not LTI system since the magnitude of the output $tx_1(t)$ is time varying.

$$x(t)=cos(2\pi t) = \frac{1}{2}(e^{j2\pi t} + e^{-j2\pi t})$$
<p> The Fourier Series representation of a signal$x(t)$ is </p>
$$ x(t)= \sum_{k=-\infty}^\infty a_k e^{jk(\frac{2\pi}{N})t} $$
<p>Thus $ x_2(t)= \sum_{k=-\infty}^\infty a_k e^{jk(\frac{2\pi}{3})t} and \: \: a_{-3}=a_3=\frac{1}{2}, \: \: a_{-2}=a_2=0, \: \: a_{-1}=a_1=0$</p>
<p> </p>
$$x_2(t)=\frac{1}{2}(e^{j3(\frac{2\pi}{3}) t} + e^{-j3(\frac{2\pi}{3}) t})$$

```python
ts = np.linspace(0, 2 * np.pi, 300)

fig, (ax1, ax2) = plt.subplots(2, 1)
fig.set_size_inches(8, 5)

# x(t) with period 1
x1 = ((1/2) * np.exp(1j*2*np.pi*ts)) + ((1/2) * np.exp(-1j*2*np.pi*ts))

ax1.plot(ts, x1.real, label='Re')
ax1.plot(ts, x1.imag, '--', label='Im')
ax1.set_title('x(t)')
ax1.set_xlabel('t')
ax1.set_ylabel('x(t)')

# x(t) with period 3
x2 = ((1/2) * np.exp(1j*3*((2*np.pi)/3)*ts)) + ((1/2) * np.exp(-1j*3*((2*np.pi)/3)*ts))

ax2.plot(ts, x2.real)
ax2.plot(ts, x2.imag, '--')
ax2.set_title('x2(t)')
ax2.set_xlabel('t')
ax2.set_ylabel('x2(t)')

ax1.legend(loc='lower right')
plt.tight_layout()
```

## 4.18

<p> Thus, the Time Domain Representation of $H(\omega)=\frac{(\sin^2(3\omega))\cos(\omega)}{\omega^2}\: \: \Longrightarrow \frac{1}{4} (\frac{2\sin(3t)}{\omega}) (\frac{2\sin(3t)}{\omega})\cos(\omega)$</p>
<p> Time domain representation: </p>
<p> $H_1(\omega)=\frac{2\sin(3\omega)}{\omega} \Longleftrightarrow 
h_1(t)= \left\{
        \begin{array}{ll}
            1 & \quad |t| < 3 \\
            0 & \quad |t| > 3
        \end{array}
    \right.$
</p>
<p> $H_2(\omega)=\frac{2\sin(3\omega)}{\omega} \Longleftrightarrow 
h_2(t)= \left\{
        \begin{array}{ll}
            1 & \quad |t| < 3 \\
            0 & \quad |t| > 3
        \end{array}
    \right.$
</p>
<p> $H_3(\omega)=\cos(\omega) = \frac{1}{2}(e^{j\omega} + e^{-j\omega}) \Longleftrightarrow h_3(t)=\frac{1}{2}\delta(t+1) + \frac{1}{2}\delta(t-1)$ </p>
<p> $H(\omega)=\frac{(\sin^2(3\omega))\cos(\omega)}{\omega^2} \Longleftrightarrow h(t)=\frac{1}{2}x(t+1) + \frac{1}{2}x(t-1)=\left\{
        \begin{array}{ll}
            0, & \quad t < -7 \\
            \frac{t+7}{8}, & \quad -7 \leq t < -5 \\
            \frac{t+6}{4}, & \quad -5 \leq t < -1 \\
            \frac{5}{4}, & \quad -1 \leq t < 1 \\
            \frac{-t+6}{4}, & \quad 1 \leq t < 5 \\
            \frac{-t+7}{8}, & \quad 5 \leq t < 7 \\
            0, & \quad t > 7
        \end{array}
    \right.$</p>

```python
ts = np.linspace(-2 * np.pi, 2 * np.pi, 600)

fig, axs = plt.subplots(4, 2)
fig.set_size_inches(12, 10)

# time domain signals
# h1(t) time domain
h1 = np.piecewise(ts, [abs(ts) < 3, abs(ts) > 3], [1, 0])

axs[0,1].plot(ts, h1)
axs[0,1].set_title('h1(t)')
axs[0,1].set_xlabel('t')
axs[0,1].set_ylabel('h1(t)')
axs[0,1].set_ylim(-0.5, 1.5)
axs[0,1].grid()

# h2(t) time domain
h2 = np.piecewise(ts, [abs(ts) < 3, abs(ts) > 3], [1, 0])

axs[1,1].plot(ts, h2)
axs[1,1].set_title('h2(t)')
axs[1,1].set_xlabel('t')
axs[1,1].set_ylabel('h2(t)')
axs[1,1].set_ylim(-0.5, 1.5)
axs[1,1].grid()

# h3(t) time domain
h3 = 1/2 * (signal.unit_impulse(600, 252) + signal.unit_impulse(600, 347))

axs[2,1].plot(ts, h3)
axs[2,1].set_title('h3(t)')
axs[2,1].set_xlabel('t')
axs[2,1].set_ylabel('h3(t)')
axs[2,1].set_xlim(-1.5, 1.5)
axs[2,1].set_ylim(-0.5, 1.0)
axs[2,1].grid()

# h(t) = h1(t) * h2(t) * h3(t) time domain
h = np.piecewise(ts, [ts<-7, (ts>=-7) & (ts<-5), (ts>=-5) & (ts<-1), (ts>=-1) & (ts<1), (ts>=1) & (ts<5), (ts>= 5) & (ts<7), ts>7], [0, lambda ts: (ts+7)/8, lambda ts: (ts+6)/4, 5/4, lambda ts: (-ts+6)/4, lambda ts: (-ts+7)/8, 0])  
    
axs[3,1].plot(ts, h)
axs[3,1].set_title('h(t)')
axs[3,1].set_xlabel('t')
axs[3,1].set_ylabel('h(t)')
axs[3,1].set_xlim(-7.0, 7.0)
axs[3,1].set_ylim(0, 1.5)
axs[3,1].grid()


# frequency domain signals
# H1(w) frequency domain
H1 = (np.sin(3*ts)) / ts

axs[0,0].plot(ts, H1)
axs[0,0].set_title('H1(\u03C9)')
axs[0,0].set_xlabel('\u03C9')
axs[0,0].set_ylabel('sin(3 \u03C9)/ \u03C9 ')
axs[0,0].set_ylim(-1.5, 4.5)
axs[0,0].grid()

# H2(w) frequency domain
H2 = (np.sin(3*ts)) / ts

axs[1,0].plot(ts, H2)
axs[1,0].set_title('H2(\u03C9)')
axs[1,0].set_xlabel('\u03C9')
axs[1,0].set_ylabel('sin(3 \u03C9)/ \u03C9 ')
axs[1,0].set_ylim(-1.5, 4.5)
axs[1,0].grid()

# H3(w) frequency domain
H3 = (np.cos(ts))

axs[2,0].plot(ts, H3)
axs[2,0].set_title('H3(\u03C9)')
axs[2,0].set_xlabel('\u03C9')
axs[2,0].set_ylabel('cos(\u03C9)')
axs[2,0].set_ylim(-1.5, 1.5)
axs[2,0].grid()

# H(w) = H1(w) * H2(w) * H3(w) frequency domain
H = H1 * H2 * H3

axs[3,0].plot(ts, H)
axs[3,0].set_title('H(\u03C9)')
axs[3,0].set_xlabel('\u03C9')
axs[3,0].set_ylabel('sin^2(3\u03C9)cos(\u03C9)/\u03C9^2')
axs[3,0].set_ylim(-1.0, 11)
axs[3,0].grid()

plt.tight_layout()
```

## 4.19

<p> $ H(j\omega) = \frac{1}{(j\omega+3)}$ </p>
<p> $ y(t) = e^{-3t}u(t) - e^{-4t}u(t) \Longleftrightarrow Y(j\omega) = \frac{1}{(j\omega+3)(j\omega+4)} $ </p>
<p> The frequency response of a system is $ \; H(j\omega) = \frac{Y(j\omega)}{X(j\omega)} \; \; thus \; \; X(j\omega) = \frac{Y(j\omega)}{H(j\omega)} $ </p>
<p> $ X(j\omega) = \frac{\frac{1}{(j\omega+3)(j\omega+4)}}{\frac{1}{(j\omega+3)}} = \frac{(j\omega+3)}{(j\omega+3)(j\omega+4)} = \frac{1}{(j\omega+3)} \Longleftrightarrow x(t) = e^{-4t}u(t) $ </p>

```python
ts = np.linspace(0, 2 * np.pi, 300)

fig, (ax1, ax2) = plt.subplots(2, 1)
fig.set_size_inches(8, 6)

# x(t)
x = np.exp(-4*ts)

ax1.plot(ts, x)
ax1.set_title('x(t)')
ax1.set_xlabel('t')
ax1.set_ylabel('x(t)')

# y(t) 
y = np.exp(-3*ts) - np.exp(-4*ts)

ax2.plot(ts, y)
ax2.set_title('y(t)')
ax2.set_xlabel('t')
ax2.set_ylabel('y(t)')

plt.tight_layout()
```

```python
# magnitude of the frequency response in decibels
def H(w):
    wc = 180
    return 1 + ((1j * w) / wc)

f = np.logspace(-1,3) # frequencies from 10**0 to 10**3

fig, (ax1, ax2) = plt.subplots(1, 2)
fig.suptitle('5.47a')
fig.set_size_inches(15, 4)

ax1.plot(f, 20*np.log10(abs(H(2*np.pi*f))))

ax1.set_xlabel('w(rad/s)')
ax1.set_ylabel('|H(w)|dB')
ax1.set_xscale('log')
ax1.set_ylim(-20.0, 40.0)
ax1.grid(True, which="both")


# angle of the frequency response in degrees
def theta(w):
    return np.arctan(w/180)

f = np.logspace(-1,3) # frequencies from 10**0 to 10**3

ax2.plot(f, (theta(2*np.pi*f)))

ax2.set_xlabel('w(rad/s)')
ax2.set_ylabel('theta(w)degrees')
ax2.set_xscale('log')
ax2.set_ylim(-np.pi / 4.0, (3.0 *np.pi) / 4.0)
ax2.grid(True, which="both")
```
